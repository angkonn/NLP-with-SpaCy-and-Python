## Standard Pipes (Components and Factories) Available from spaCy
SpaCy is much more than an NLP framework. It is also a way of designing and implementing complex pipelines. A pipeline is a sequence of pipes, or actors on data, that make alterations to the data or extract information from it. In some cases, later pipes require the output from earlier pipes. In other cases, a pipe can exist entirely on its own. 

Here, we see an input, in this case a sentence, enter the pipeline from the left. Two pipes are activated on this, a rules-based named entity recognizer known as an EntityRuler which finds entities and an EntityLinker pipe that identifies what entity that is to perform toponym resolution. The sentence is then outputted with the sentence and the entities annotated. At this point, we could use the doc.ents feature to find the entities in our sentence. In spaCy, you will often use pipelines that are more sophisticated than this. You will specifically use a Tok2Vec input layer to vectorize your input sentence. This will allow machine learning pipes to make predictions.

## How to Add Pipes
In most cases, you will use an off-the-shelf spaCy model. In some cases, however, an off-the-shelf model will not fill your needs or will perform a specific task very slowly. A good example of this is sentence tokenization. Imagine if you had a document that was around 1 million sentences long. Even if you used the small English model, your model would take a long time to process those 1 million sentences and separate them. In this instance, you would want to make a blank English model and simply add the Sentencizer to it. The reason is because each pipe in a pipeline will be activated (unless specified) and that means that each pipe from Dependency Parser to named entity recognition will be performed on your data. This is a serious waste of computational resources and time. The small model may take hours to achieve this task. By creating a blank model and simply adding a Sentencizer to it, you can reduce this time to merely minutes.

To demonstrate this process, lets first create a blank model.

```python
nlp = spacy.blank("en")
```

```python
nlp.add_pipe("sentencizer")
```

```python
nlp.analyze_pipes()
```

## Output Explanation

The output is a dictionary that provides a detailed breakdown of the pipeline components in the `en_core_web_sm` model. Let's analyze the key parts:

### 1. summary:
This section lists each component in the pipeline and its properties.

**tok2vec**:
- **Purpose**: Converts tokens (words) into vector representations (embeddings) that other components use.
- **Assigns**: `doc.tensor` (adds a tensor representation to the Doc object).
- **Requires**: `[]` (no dependencies on other components).
- **Scores**: `[]` (no evaluation scores, as it's a preprocessing step).
- **Retokenizes**: `False` (doesn't modify token boundaries).

**tagger**:
- **Purpose**: Assigns part-of-speech (POS) tags to each token (e.g., noun, verb, adjective).
- **Assigns**: `token.tag` (adds POS tags to tokens).
- **Requires**: `[]` (no dependencies, but often uses tok2vec embeddings).
- **Scores**: `['tag_acc', 'pos_acc', 'tag_micro_p', 'tag_micro_r', 'tag_micro_f']` (metrics like tag accuracy, precision, recall, and F1-score for POS tagging).
- **Retokenizes**: `False` (doesn't change tokenization).

**parser**:
- **Purpose**: Performs dependency parsing to identify syntactic relationships between tokens (e.g., subject, object).
- **Assigns**:
  - `token.dep` (dependency labels, e.g., "nsubj" for nominal subject).
  - `token.head` (the head token in the dependency tree).
  - `token.is_sent_start` (marks the start of sentences).
  - `doc.sents` (defines sentence boundaries in the Doc).
- **Requires**: `[]` (no strict dependencies, but uses tok2vec).
- **Scores**: `['dep_uas', 'dep_las', 'dep_las_per_type', 'sents_p', 'sents_r', 'sents_f']` (metrics for dependency parsing accuracy and sentence segmentation).
- **Retokenizes**: `False`.

**attribute_ruler**:
- **Purpose**: Applies rule-based mappings to token attributes (e.g., for custom POS or lemma assignments).
- **Assigns**: `[]` (in this model, it may not assign anything by default).
- **Requires**: `[]`.
- **Scores**: `[]`.
- **Retokenizes**: `False`.

Other components (like `ner`) would follow a similar structure, but the output is truncated here. For example:

**ner** (Named Entity Recognition):
- **Assigns**: `doc.ents`, `token.ent_type` (identifies entities like PERSON, ORG, and their types).
- **Requires**: `[]`.
- **Scores**: Metrics like precision, recall, and F1 for entity recognition.

### 2. Attribute-to-Component Mapping:
The second part of the output (e.g., `'token.ent_type': {'assigns': ['ner'], 'requires': []}`) shows which pipeline components assign or require specific attributes:

- `token.ent_type`: Assigned by the `ner` component (indicates the type of named entity, e.g., "PERSON").
- `token.tag`: Assigned by the `tagger` component (POS tag).
- `doc.ents`: Assigned by the `ner` component (list of entities in the document).
- `token.dep`: Assigned by the `parser` component (dependency label).
- `doc.sents`: Assigned by the `parser` component (sentence spans).

Each attribute lists:
- `assigns`: Which component(s) set this attribute.
- `requires`: Any components required to compute this attribute (empty in this case, as `en_core_web_sm` components are independent).

### What This Tells Us
- The `en_core_web_sm` pipeline includes components like `tok2vec`, `tagger`, `parser`, `attribute_ruler`, and likely `ner` (based on the attribute mapping).
- Each component adds specific linguistic annotations to the Doc object (e.g., POS tags, dependency relations, entities).
- The pipeline is modular, with no components requiring others (`requires: []`), but they often rely on `tok2vec` for embeddings.
- The scores indicate how the model's performance can be evaluated (e.g., accuracy for POS tags or dependency parsing).
- No component retokenizes the text, meaning the token boundaries remain fixed after initial tokenization.

### Practical Use
The `analyze_pipes()` output is useful for:
- Understanding what each component in the pipeline does.
- Checking which attributes (e.g., `token.tag`, `doc.ents`) are available for your analysis.
- Debugging or customizing the pipeline (e.g., disabling components with `nlp2 = spacy.load("en_core_web_sm", disable=["parser"])`).
- Evaluating model performance using the listed scores.
